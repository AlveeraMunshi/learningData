#### Party expense exercise

Suppose that you are responsible for buying pizza and soda for your employer's annual company party. You must buy entire pizza pies and cases of soda (boxes of soda cans). You cannot buy individual slices of pizza or cans of soda, nor can you buy fractional numbers of pizza pies or soda cases.

Write a function in Python to compute the total cost of the pizza and soda you will need to buy. Do not buy more than you need to in order to supply each partygoer with the desired number of pizza slices and soda cans.

Your function should take the following parameters:

1. the number of people attending the party (default 5)
2. the number of slices of pizza each person should be able to eat (default 2)
3. the number of cans of soda each person should be able to drink (default 1)
4. the cost of a pizza pie (default \$12)
5. the number of slices in a pizza pie (default 8)
6. the cost of a case of soda (default \$7.50)
7. the number of cans in a case of soda (default 12)

```python
import math

def calculate_party_cost(num_people=5, 
                         slices_per_person=2, 
                         cans_per_person=1, 
                         cost_per_pizza=12, 
                         slices_per_pizza=8, 
                         cost_per_soda_case=7.50, 
                         cans_per_case=12):
    # Calculate total slices of pizza needed
    total_slices_needed = num_people * slices_per_person
    # Calculate total pizza pies needed (round up to the nearest whole pie)
    total_pizza_pies = math.ceil(total_slices_needed / slices_per_pizza)
    
    # Calculate total cans of soda needed
    total_cans_needed = num_people * cans_per_person
    # Calculate total cases of soda needed (round up to the nearest whole case)
    total_soda_cases = math.ceil(total_cans_needed / cans_per_case)
    
    # Calculate the total cost for pizzas and sodas
    total_cost_pizzas = total_pizza_pies * cost_per_pizza
    total_cost_sodas = total_soda_cases * cost_per_soda_case
    
    # Calculate the total cost
    total_cost = total_cost_pizzas + total_cost_sodas
    
    return total_cost

# Example usage:
total_cost = calculate_party_cost()
print(f"The total cost for the party is: ${total_cost:.2f}")
```

```python
import math

def calculate_party_cost(num_people, 
                         slices_per_person, 
                         cans_per_person, 
                         cost_per_pizza=12, 
                         slices_per_pizza=8, 
                         cost_per_soda_case=7.50, 
                         cans_per_case=12):
    # Calculate total slices of pizza needed
    total_slices_needed = num_people * slices_per_person
    # Calculate total pizza pies needed (round up to the nearest whole pie)
    total_pizza_pies = math.ceil(total_slices_needed / slices_per_pizza)
    
    # Calculate total cans of soda needed
    total_cans_needed = num_people * cans_per_person
    # Calculate total cases of soda needed (round up to the nearest whole case)
    total_soda_cases = math.ceil(total_cans_needed / cans_per_case)
    
    # Calculate the total cost for pizzas and sodas
    total_cost_pizzas = total_pizza_pies * cost_per_pizza
    total_cost_sodas = total_soda_cases * cost_per_soda_case
    
    # Calculate the total cost
    total_cost = total_cost_pizzas + total_cost_sodas
    
    return total_pizza_pies, total_soda_cases, total_cost

# Example usage:
num_people = 10
slices_per_person = 3
cans_per_person = 2
pizza_pies, soda_cases, total_cost = calculate_party_cost(num_people, slices_per_person, cans_per_person)
print(f"The total number of pizza pies needed is: {pizza_pies}")
print(f"The total number of soda cases needed is: {soda_cases}")
print(f"The total cost for the party is: ${total_cost:.2f}")
```
